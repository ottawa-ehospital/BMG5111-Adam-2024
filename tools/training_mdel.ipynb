```python
df = df.drop(['Initial_EDSS', 'Final_EDSS'], axis=1)
df = df.fillna(df.mean(), axis=0)
```

```python
data = df.copy()
```

```python
gender = {1:'Male', 2: 'Female'}
breast_feeding = {1: 'yes', 2:'no', 3:'unknown'}
varicella = {1 : 'positive', 2: 'negative', 3: 'unknown'}
group = {1: 'CDMS' , 2: 'Non-CDMS' }

data['Gender'] = data['Gender'].map(gender)
data['Breastfeeding'] = data['Breastfeeding'].map(breast_feeding)
data['Varicella'] = data['Varicella'].map(varicella)
data['group'] = data['group'].map(group)
```

```python
import matplotlib.pyplot as plt
a = data['group'].value_counts()
plt.pie(a.values, labels = a.index, autopct='%.2f')
```

```python
col = ['Gender', 'Varicella', 'Initial_Symptom', 'LLSSEP','ULSSEP', 'VEP', 'BAEP', 'Periventricular_MRI', 'Cortical_MRI', 'Infratentorial_MRI', 'Spinal_Cord_MRI']
X = df[col]
X.head()
```

```python
y = df['group']%2
y.head()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV, StratifiedKFold

X_train, X_valid, y_train, y_valid = train_test_split(X,y,stratify=y,train_size=0.8,test_size=0.2,random_state=0)
X_test = X_valid
X_valid.to_csv('../data/test_data.csv', index=False)
# X_valid.to_csv('test_data.csv', index=False)
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.naive_bayes import GaussianNB

# Classifiers
classifiers = {
    "LogisticRegression" : LogisticRegression(random_state=0),
    "KNN" : KNeighborsClassifier(),
    "SVC" : SVC(random_state=0, probability=True),
    "RandomForest" : RandomForestClassifier(random_state=0),
    "XGBoost" : XGBClassifier(random_state=0, use_label_encoder=False, eval_metric='logloss'), # XGBoost takes too long
    "LGBM" : LGBMClassifier(random_state=0),
    "CatBoost" : CatBoostClassifier(random_state=0, verbose=False),
    "NaiveBayes": GaussianNB()
}

# Grids for grid search
LR_grid = {'penalty': ['l1','l2'],
           'C': [0.25, 0.5, 0.75, 1, 1.25, 1.5],
           'max_iter': [50, 100, 150]}

KNN_grid = {'n_neighbors': [3, 5, 7, 9],
            'p': [1, 2]}

SVC_grid = {'C': [0.25, 0.5, 0.75, 1, 1.25, 1.5],
            'kernel': ['linear', 'rbf'],
            'gamma': ['scale', 'auto']}

RF_grid = {'n_estimators': [50, 100, 150, 200, 250, 300],
        'max_depth': [4, 6, 8, 10, 12]}

boosted_grid = {'n_estimators': [50, 100, 150, 200],
        'max_depth': [4, 8, 12],
        'learning_rate': [0.05, 0.1, 0.15]}

NB_grid={'var_smoothing': [1e-10, 1e-9, 1e-8, 1e-7]}

# Dictionary of all grids
grid = {
    "LogisticRegression" : LR_grid,
    "KNN" : KNN_grid,
    "SVC" : SVC_grid,
    "RandomForest" : RF_grid,
    "XGBoost" : boosted_grid,
    "LGBM" : boosted_grid,
    "CatBoost" : boosted_grid,
    "NaiveBayes": NB_grid
}

import numpy as np # linear algebra
import time
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import recall_score, f1_score, roc_auc_score
i=0
clf_best_params=classifiers.copy()
valid_scores=pd.DataFrame({'Classifer':classifiers.keys(), 'Validation accuracy': np.zeros(len(classifiers)),
                           'recall_score': np.zeros(len(classifiers)),'f1_score': np.zeros(len(classifiers)),
                           'roc_auc_score': np.zeros(len(classifiers)),'Training time': np.zeros(len(classifiers))})
for key, classifier in classifiers.items():
    start = time.time()
    clf = GridSearchCV(estimator=classifier, param_grid=grid[key], n_jobs=-1, cv=None)

    # Train and score
    clf.fit(X_train, y_train)
    
    valid_scores.iloc[i,1]=clf.score(X_valid, y_valid)
    # 使用最佳参数的模型在验证集上进行预测
    y_pred = clf.predict(X_valid)  # 对于召回率和 F1 分数
    y_probs = clf.predict_proba(X_valid)[:, 1]  # 对于 AUC-ROC 分数，需要预测概率

    # 计算评分
    recall = recall_score(y_valid, y_pred)
    f1 = f1_score(y_valid, y_pred)
    roc_auc = roc_auc_score(y_valid, y_probs)

    # 将评分保存到 valid_scores DataFrame
    valid_scores.iloc[i, 2] = recall
    valid_scores.iloc[i, 3] = f1
    valid_scores.iloc[i, 4] = roc_auc
    

    # Save trained model
    clf_best_params[key]=clf.best_params_
    
    # Print iteration and training time
    stop = time.time()
    valid_scores.iloc[i,5]=np.round((stop - start)/60, 2)
    
    print('Model:', key)
    print('Training time (mins):', valid_scores.iloc[i,2])
    print('')
    i+=1
```

```python
# {'LogisticRegression': {'C': 0.25, 'max_iter': 50, 'penalty': 'l2'},
#  'KNN': {'n_neighbors': 7, 'p': 1},
#  'SVC': {'C': 1, 'gamma': 'scale', 'kernel': 'linear'},
#  'RandomForest': {'max_depth': 4, 'n_estimators': 250},
#  'XGBoost': {'learning_rate': 0.05, 'max_depth': 8, 'n_estimators': 50},
#  'LGBM': {'learning_rate': 0.05, 'max_depth': 8, 'n_estimators': 50},
#  'CatBoost': {'learning_rate': 0.05, 'max_depth': 8, 'n_estimators': 50},
#  'NaiveBayes': {'var_smoothing': 1e-10}}
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.naive_bayes import GaussianNB


# 设置最佳参数
best_params = {
    'LogisticRegression': {'C': 0.25, 'max_iter': 50, 'penalty': 'l2'},
    'KNN': {'n_neighbors': 7, 'p': 1},
    'SVC': {'C': 1, 'gamma': 'scale', 'kernel': 'linear'},
    'RandomForest': {'max_depth': 4, 'n_estimators': 250},
    'XGBoost': {'learning_rate': 0.05, 'max_depth': 8, 'n_estimators': 50},
    'LGBM': {'learning_rate': 0.05, 'max_depth': 8, 'n_estimators': 50},
    'CatBoost': {'learning_rate': 0.05, 'max_depth': 8, 'n_estimators': 50},
    'NaiveBayes': {'var_smoothing': 1e-10}
}

from sklearn.ensemble import VotingClassifier

# 创建模型实例并应用最佳参数
log_clf = LogisticRegression(**best_params['LogisticRegression'])
knn_clf = KNeighborsClassifier(**best_params['KNN'])
svc_clf = SVC(**best_params['SVC'], probability=True)  # 启用概率估计以进行软投票
rf_clf = RandomForestClassifier(**best_params['RandomForest'])
xgb_clf = XGBClassifier(**best_params['XGBoost'], use_label_encoder=False, eval_metric='logloss')  # 避免XGBoost特定的警告
lgbm_clf = LGBMClassifier(**best_params['LGBM'])
cat_clf = CatBoostClassifier(**best_params['CatBoost'], verbose=0)  # 避免过多日志输出
gnb_clf = GaussianNB(**best_params['NaiveBayes'])


voting_clf = VotingClassifier(
    estimators=[
        ('lr', log_clf),
        ('knn', knn_clf),
        ('svc', svc_clf),
        ('rf', rf_clf),
        ('xgb', xgb_clf),
        ('lgbm', lgbm_clf),
        ('cat', cat_clf),
        ('gnb', gnb_clf)
    ],
    voting='hard'  # 使用软投票
)

# 使用训练数据拟合投票分类器
voting_clf.fit(X_train, y_train)

# 使用验证集评估模型性能
score = voting_clf.score(X_valid, y_valid)
print("Voting Classifier performance:", score)
y_pred = voting_clf.predict(X_valid)  # 对于召回率和 F1 分数
# y_probs = voting_clf.predict_proba(X_valid)[:, 1]  # 对于 AUC-ROC 分数，需要预测概率

recall = recall_score(y_valid, y_pred)
f1 = f1_score(y_valid, y_pred)
# roc_auc = roc_auc_score(y_valid, y_probs)
print(f"recall -> {recall}")
print(f"f1 -> {f1}")
# print(f"roc_auc -> {roc_auc}")
from sklearn.metrics import recall_score, f1_score, roc_auc_score
```
